#Question 1: What is the difference between descriptive statistics and inferential statistics? Explain with examples.

Descriptive statistics summarize and describe the features of a dataset, providing a clear overview of the data. Inferential statistics, on the other hand, make inferences and draw conclusions about a larger population based on a sample of that population.

**Descriptive Statistics:**
- Definition: These statistics describe, summarize, and organize characteristics of a dataset. They provide a concise summary of the data without attempting to generalize beyond the data itself.

**Purpose:** To describe the main features of the data.

**Examples:**
  - Calculating the mean, median, or mode of a student's test scores to understand the average performance.
  - Creating a histogram to visualize the distribution of ages in a survey group.
  - Determining the standard deviation of salaries in a company to understand the spread of income.

**Inferential Statistics:**
- **Definition:** These statistics use data from a sample to make predictions, generalizations, or inferences about a larger population. They involve probability to make conclusions beyond the observed data.
- **Purpose:** To make predictions and test hypotheses about a population based on sample data.
- **Examples:**
  - A researcher takes a sample of 100 students from a university of 10,000 to determine if a new teaching method improves test scores. The results from the sample are then used to infer if the method would be effective for all students at the university.
  - A political pollster surveys 1,000 likely voters to predict the outcome of an election for an entire country.
  - Medical researchers conducting a clinical trial on a sample of patients to determine the effectiveness of a new drug on the general population with a specific disease.

#Question 2: What is sampling in statistics? Explain the differences between random and stratified sampling.

##What is Sampling in Statistics?

**Sampling** is a process used in statistical analysis where a predetermined number of observations are taken from a larger population. The idea is that instead of studying an entire population (which can be too large, costly, or time-consuming), we select a representative subset (a "sample") and study that subset. The insights gained from the sample are then used to make inferences or predictions about the entire population.

##Why is Sampling Important?

- **Feasibility:** Studying an entire population is often impractical or impossible.
- **Cost-effectiveness:** Collecting data from a sample is significantly cheaper.
- **Time-efficiency:** It saves a lot of time compared to surveying every single unit in a population.
- **Accuracy:** If done correctly, sampling can provide accurate and reliable results, sometimes even more accurate than a full census due to fewer errors in data collection and processing.

##Differences Between Random Sampling and Stratified Sampling

Both random sampling and stratified sampling are methods of probability sampling, meaning every element in the population has a known, non-zero chance of being selected. However, they differ in how they approach the selection process.

##Random Sampling (Simple Random Sampling)
Definition: Simple Random Sampling (SRS) is a method where every individual (or unit) in the population has an equal and independent chance of being selected for the sample. Think of it like drawing names out of a hat. Each name has the same probability of being chosen, and the selection of one name doesn't affect the selection of another.

**How it works:**
1. Define the population clearly.
2. Assign a unique identifier (e.g., a number) to each individual in the population.
3. Use a random process (like a random number generator) to select the required number of individuals for the sample.

**Example:**
- A high school wants to survey its students about cafeteria food. There are 1000 students. Using simple random sampling, they put all 1000 student names into a list and randomly select 100 names. Each student has a 1/10 chance of being selected, irrespective of their grade level or gender.
##Stratified Sampling
**Definition:** Stratified sampling is a method that involves dividing the population into smaller subgroups (called "strata" or "stratum" for singular) based on shared characteristics (e.g., age group, gender, income level, education level). Then, a simple random sample is drawn from each stratum. This ensures that specific subgroups are proportionally (or disproportionately, if desired for specific analysis) represented in the overall sample.

**How it works:**

1. Divide the population into mutually exclusive and collectively exhaustive strata based on relevant characteristics.
2. Determine the sample size for each stratum. This can be proportional (e.g., if a stratum makes up 20% of the population, it makes up 20% of the sample) or disproportionate (if a smaller stratum is of particular interest and needs more representation).
3. Perform simple random sampling within each stratum to select the individuals.
4. Combine the samples from all strata to form the final sample.

**Example:**

The same high school wants to survey its students about cafeteria food, but this time, they want to ensure that each grade level (9th, 10th, 11th, 12th) is adequately represented. They decide to use stratified sampling.
  - **Strata:** They divide the 1000 students into four strata: 9th Grade (300 students), 10th Grade (250 students), 11th Grade (250 students), 12th Grade (200 students).
  - **Proportional Sampling:** If they want a sample of 100 students, they would randomly select:
    - 30 students from 9th Grade (30% of total sample, matching 30% of population)
    - 25 students from 10th Grade (25% of total sample)
    - 25 students from 11th Grade (25% of total sample)
    - 20 students from 12th Grade (20% of total sample)
  - They would perform a simple random sample within each grade level to select the students.

### Summary of Differences:

| Feature             | Random Sampling (Simple Random)                   | Stratified Sampling                                     |
| :------------------ | :------------------------------------------------ | :------------------------------------------------------ |
| **Population Division** | No division; treated as one homogeneous group.    | Divided into distinct, non-overlapping subgroups (strata). |
| **Selection Process** | Randomly selects individuals from the entire population. | Randomly selects individuals from *within each stratum*.    |
| **Representation**  | Relies on chance for representativeness; may miss small subgroups. | Guarantees representation of all strata, proportional or not. |
| **Homogeneity**     | Assumes population is largely homogeneous or that heterogeneity doesn't matter. | Assumes heterogeneity across strata, homogeneity within strata. |
| **Complexity**      | Simpler to implement.                             | More complex; requires more information about the population. |
| **Precision**       | Lower precision for population estimates if significant subgroup variation exists. | Higher precision, especially when strata are well-defined and distinct. |
| **Use Case**        | When population is relatively homogeneous or no specific subgroup comparison is needed. | When population is heterogeneous and representation of specific subgroups is crucial, or for subgroup comparisons. |

#Question 3: Define mean, median, and mode. Explain why these measures of central tendency are important.


###Measures of Central Tendency: Mean, Median, and Mode
Measures of central tendency are single values that attempt to describe a set of data by identifying the central position within that set of data. They are often called "averages." There are three main measures: the mean, the median, and the mode.

**1. Mean (Arithmetic Mean)**

**Definition:** The mean, often simply called the "average," is calculated by summing all the values in a dataset and then dividing by the number of values in the dataset.

**Example:** If you have test scores: `[85, 90, 78, 92, 88]` Mean $= (85 + 90 + 78 + 92 + 88) / 5 = 433 / 5 = 86.6$$= (85 + 90 + 78 + 92 + 88) / 5 = 433 / 5 = 86.6$

**Why it's Important:**
- **Widely Understood:** It's the most commonly used measure of central tendency and is intuitively understood by most people.
- **Uses All Data Points:** Every value in the dataset contributes to the calculation of the mean, making it a good summary of the entire dataset.
- **Foundation for Other Statistics:** The mean is a fundamental building block for many other statistical concepts and calculations, such as variance, standard deviation, and regression analysis.
- **Mathematical Properties:** It has desirable mathematical properties that make it suitable for various statistical analyses and hypothesis testing.

**Limitation:** The mean is highly sensitive to outliers (extreme values). A single very high or very low value can significantly pull the mean in that direction, making it less representative of the typical value in skewed distributions.

**2. Median**

**Definition:** The median is the middle value in a dataset when the data is arranged in numerical order. It divides the dataset into two equal halves, with half the values falling below the median and half falling above it.

**How to find it:**

1. Arrange all data points in ascending or descending order.
2. If the number of data points ($n$$n$) is odd, the median is the middle value.
3. If the number of data points ($n$$n$) is even, the median is the average of the two middle values.

**Example:**

- **Odd number of values:** [78, 85, 88, 90, 92] (sorted) Median $= 88$
- **Even number of values:** [78, 85, 88, 90, 92, 95] (sorted) Median $= (88 + 90) / 2 = 89$

**Why it's Important:**
- **Robust to Outliers:** Unlike the mean, the median is not affected by extreme outliers. This makes it a more reliable measure of central tendency for skewed distributions (e.g., income data, property prices) where a few very high or low values could distort the mean.
- **Represents the "Typical" Value:** It provides a good sense of the "typical" value when the data is not symmetrically distributed, as it represents the point where half the data lies above and half lies below.
- **Easy to Understand:** It's relatively easy to interpret as the midpoint of the data.

**3. Mode**

**Definition:** The mode is the value that appears most frequently in a dataset. A dataset can have one mode (unimodal), multiple modes (multimodal), or no mode if all values appear with the same frequency.

**Example:**

- `[85, 90, 78, 92, 88, 90, 85, 90]` Mode $= 90$ (appears 3 times, more than any other value)
- `[1, 2, 3, 3, 4, 4, 5]` Modes $= 3, 4$ (bimodal)
- `[10, 20, 30, 40]` No mode (all values appear once)

**Why it's Important:**
- **Applicable to Categorical Data:** The mode is the only measure of central tendency that can be used with categorical (non-numerical) data (e.g., favorite color, most popular car brand).
- **Identifies Popular/Frequent Values:** It directly tells us which value or category is the most common or popular, which is valuable in market research, voting patterns, or product preferences.
- **Indicates Peaks in Distribution:** In a frequency distribution or histogram, the mode corresponds to the peak(s) of the distribution, indicating where the data is most concentrated.
###Overall Importance of Measures of Central Tendency
These three measures are critical in statistics and data science for several reasons:

1. **Summarize Data:** They provide a quick, single-value summary that helps us understand the typical or central value of a large dataset, making data more digestible.
2. **Facilitate Comparison:** They allow for easy comparison between different datasets or groups. For instance, comparing the average sales across different regions.
3. **Reveal Data Characteristics:** By comparing the mean, median, and mode, we can gain insights into the shape and skewness of a distribution:
If Mean ≈ Median ≈ Mode: The distribution is likely symmetrical (e.g., a normal distribution).
  - If **Mean > Median > Mode:** The distribution is positively (right) skewed, indicating a long tail towards higher values.
  - If **Mean < Median < Mode:** The distribution is negatively (left) skewed, indicating a long tail towards lower values.
4. **Basis for Further Analysis:** They form the foundation for more advanced statistical analyses, such as inferential statistics, where these central values are used to make predictions or draw conclusions about populations.
5. **Decision Making:** In fields like business, economics, and healthcare, these measures are used to make informed decisions – whether it's setting product prices, analyzing market trends, or understanding patient demographics.

#Question 4: Explain skewness and kurtosis. What does a positive skew imply about the data?

**Skewness**

**Definition:** Skewness is a measure of the asymmetry of the probability distribution of a real-valued random variable about its mean. In simpler terms, it tells us if the data is concentrated on one side and has a longer 'tail' on the other.
- **Symmetrical Distribution:** If a distribution is perfectly symmetrical (like a normal distribution or bell curve), its skewness is zero. The mean, median, and mode are typically all in the same location.
- **Positive Skew (Right-skewed):** The tail on the right side of the distribution is longer or fatter than the tail on the left side. This means there are a few unusually large values (outliers) that pull the mean towards the right (higher values). In a positively skewed distribution, the mode is typically less than the median, which is less than the mean (Mode < Median < Mean).
- **Negative Skew (Left-skewed):** The tail on the left side of the distribution is longer or fatter than the tail on the right side. This indicates the presence of a few unusually small values (outliers) that pull the mean towards the left (lower values). In a negatively skewed distribution, the mean is typically less than the median, which is less than the mode (Mean < Median < Mode).

**What a Positive Skew Implies about the Data:**

When a data distribution exhibits **positive skewness**, it implies the following:

1. **Tail on the Right:** The right side of the distribution extends further out, indicating the presence of some higher values that are far from the bulk of the data.
2. **Concentration of Data on the Left:** Most of the data points are clustered towards the lower end of the value range.
3. **Mean > Median > Mode:** The mean will be pulled towards the right (higher values) by the extreme high values, making it greater than the median. The mode, representing the most frequent value, will be at the peak of the distribution, which is usually to the left.
4. **Examples:** Common examples include income distribution (most people earn less, a few earn a lot), housing prices (most houses are within a certain range, but a few mansions inflate the average), and waiting times (most waits are short, but a few are very long).
5. **Practical Interpretation:** If you're looking at a positively skewed dataset, it tells you that the typical value (median or mode) is lower than what the simple average (mean) might suggest. There are more observations at the lower end, with fewer, but very high, observations stretching out the tail.

**Kurtosis**

**Definition:** Kurtosis is a measure of the "tailedness" of the probability distribution of a real-valued random variable. It characterizes the shape of the probability distribution, specifically its peak (peakedness) and the heaviness of its tails relative to a normal distribution.

- **Mesokurtic:** A distribution with kurtosis similar to that of a normal distribution. It has a kurtosis value of 0 when calculated using excess kurtosis (which subtracts 3 from the raw kurtosis, making a normal distribution's excess kurtosis 0).
- **Leptokurtic:** A distribution with positive excess kurtosis. It has a sharper peak and heavier tails than a normal distribution. This means more outliers (extreme values) are present than in a normal distribution.
  - **Implication:** Higher probability of extreme events or observations.
- **Platykurtic:** A distribution with negative excess kurtosis. It has a flatter peak and lighter tails than a normal distribution. This means fewer outliers (extreme values) are present compared to a normal distribution.
  - **Implication:** Lower probability of extreme events or observations, with values more spread out across the distribution.

**Why Skewness and Kurtosis are Important:**
- **Understanding Data Shape:** They provide crucial information about the shape of a distribution that goes beyond just central tendency and variability. This understanding is vital for selecting appropriate statistical models and tests.
- **Detecting Outliers:** Leptokurtic distributions (high kurtosis) signal the presence of more outliers, which might require special handling or investigation.
- **Risk Assessment:** In finance, leptokurtic distributions of returns indicate higher chances of extreme gains or losses, which is critical for risk management.
- **Model Assumptions:** Many statistical models (e.g., linear regression, ANOVA) assume that data is normally distributed (zero skewness, mesokurtic). Deviations indicated by skewness and kurtosis might necessitate data transformations or the use of non-parametric methods.
- **Better Data Interpretation:** They help in giving a complete picture of the data, preventing misinterpretations that might arise from only looking at the mean or standard deviation. For instance, two datasets could have the same mean and standard deviation but wildly different shapes due to skewness and kurtosis.

#Question 5: Implement a Python program to compute the mean, median, and mode of a given list of numbers.
- numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

```python
import statistics
from collections import Counter

numbers = [12, 15, 12, 18, 19, 12, 20, 22, 19, 19, 24, 24, 24, 26, 28]

# Calculate Mean
mean_value = statistics.mean(numbers)

# Calculate Median
median_value = statistics.median(numbers)

# Calculate Mode
# Counter returns a dictionary-like object with elements and their counts
counts = Counter(numbers)

# Find the highest frequency
max_count = max(counts.values())

# Find all elements that have this highest frequency
# A dataset can have one mode, multiple modes (multimodal), or no mode
mode_values = [key for key, value in counts.items() if value == max_count]

print(f"Given list of numbers: {numbers}")
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: {mode_values}")
```

#Question 6: Compute the covariance and correlation coefficient between the following two datasets provided as lists in Python:
- list_x = [10, 20, 30, 40, 50]
- list_y = [15, 25, 35, 45, 60]

```python
import numpy as np

list_x = [10, 20, 30, 40, 50]
list_y = [15, 25, 35, 45, 60]

# Convert lists to NumPy arrays
array_x = np.array(list_x)
array_y = np.array(list_y)

# Compute Covariance
# The result is a covariance matrix. We are interested in the off-diagonal element (e.g., [0, 1] or [1, 0])
# which represents the covariance between x and y.
covariance_matrix = np.cov(array_x, array_y)
covariance_xy = covariance_matrix[0, 1]

# Compute Correlation Coefficient
# The result is a correlation matrix. We are interested in the off-diagonal element (e.g., [0, 1] or [1, 0])
# which represents the correlation coefficient between x and y.
correlation_matrix = np.corrcoef(array_x, array_y)
correlation_xy = correlation_matrix[0, 1]

print(f"List X: {list_x}")
print(f"List Y: {list_y}")
print(f"\nCovariance between X and Y: {covariance_xy}")
print(f"Correlation Coefficient between X and Y: {correlation_xy}")
```

#Question 7: Write a Python script to draw a boxplot for the following numeric list and identify its outliers. Explain the result:
- data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]


```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

data = [12, 14, 14, 15, 18, 19, 19, 21, 22, 22, 23, 23, 24, 26, 29, 35]

# Calculate quartiles and IQR to identify outliers programmatically
Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1

# Define outlier bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers
outliers = [x for x in data if x < lower_bound or x > upper_bound]

# Create the boxplot
plt.figure(figsize=(8, 6))
sns.boxplot(y=data, color='skyblue')
sns.stripplot(y=data, color='red', jitter=0.1, size=8, alpha=0.7) # Optional: show individual data points

plt.title('Boxplot of Data Distribution')
plt.ylabel('Values')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

print(f"Given data: {data}")
print(f"Q1 (25th percentile): {Q1}")
print(f"Q3 (75th percentile): {Q3}")
print(f"IQR (Interquartile Range): {IQR}")
print(f"Lower Bound for Outliers: {lower_bound}")
print(f"Upper Bound for Outliers: {upper_bound}")
print(f"Identified Outliers: {outliers}")

# Explanation of the results
explanation = "\nExplanation of the Boxplot and Outliers:\n"
explanation += "A boxplot visually represents the distribution of numerical data through its quartiles. The box itself spans from the first quartile (Q1) to the third quartile (Q3), with a line inside indicating the median (Q2). The 'whiskers' extend from the box to the minimum and maximum values within 1.5 times the Interquartile Range (IQR) from Q1 and Q3, respectively.\n\n"
explanation += f"For this dataset:\n"
explanation += f"- **Median (Q2):** The line inside the box is at {np.median(data)}. This means half of the data points are below this value and half are above.\n"
explanation += f"- **First Quartile (Q1):** The bottom of the box is at {Q1}. 25% of the data falls below this value.\n"
explanation += f"- **Third Quartile (Q3):** The top of the box is at {Q3}. 75% of the data falls below this value.\n"
explanation += f"- **Interquartile Range (IQR):** The height of the box is {IQR} (Q3 - Q1). This represents the middle 50% of the data.\n"
explanation += f"- **Whiskers:** The whiskers extend from Q1 to the lowest data point not considered an outlier (which is {min(data) if not outliers or min(data) > lower_bound else 'the lowest non-outlier value'}) and from Q3 to the highest data point not considered an outlier (which is {max(data) if not outliers or max(data) < upper_bound else 'the highest non-outlier value'}).\n"

if outliers:
    explanation += f"- **Outliers:** Data points outside the whiskers are considered outliers. Based on the 1.5 * IQR rule, any values below {lower_bound} or above {upper_bound} are outliers. In this dataset, the identified outlier(s) is/are: {outliers}. These are typically plotted as individual points beyond the whiskers on the boxplot.\n"
else:
    explanation += "- **Outliers:** There are no outliers identified in this dataset based on the 1.5 * IQR rule.\n"

print(explanation)
```

#Question 8: You are working as a data analyst in an e-commerce company. The marketing team wants to know if there is a relationship between advertising spend and daily sales.
- Explain how you would use covariance and correlation to explore this relationship.
- Write Python code to compute the correlation between the two lists:
- advertising_spend = [200, 250, 300, 400, 500]
- daily_sales = [2200, 2450, 2750, 3200, 4000]

**Exploring the Relationship: Covariance and Correlation**

**1. Using Covariance**
- **What it is:** Covariance measures the direction of the linear relationship between two variables. In our case, it would tell us if an increase in advertising spend is generally associated with an increase (or decrease) in daily sales.
- **Interpretation in this context:**
  - **Positive Covariance:** If the covariance is positive, it suggests that as advertising spend increases, daily sales tend to increase. This would be a favorable indication for the marketing team, hinting that their ad campaigns are effective.
  - **Negative Covariance:** If the covariance is negative, it implies that as advertising spend increases, daily sales tend to decrease. This would be a red flag, indicating a potential problem with the advertising strategy or market conditions.
  - **Near Zero Covariance:** A covariance close to zero suggests no clear linear relationship between the two variables. This doesn't necessarily mean there's no relationship at all (there could be a non-linear one), but it indicates no straightforward upward or downward trend.
- **Limitation:** The magnitude of covariance is hard to interpret directly because it's scale-dependent. A large covariance could mean a strong relationship, or it could simply mean the values of the variables themselves are large. For instance, a covariance of 200 between sales and spend means something different if sales are in thousands versus millions.

**2. Using Correlation Coefficient (Pearson's r)**
- **What it is:** The correlation coefficient (most commonly Pearson's product-moment correlation coefficient) takes covariance a step further. It not only tells us the direction of the linear relationship but also its strength, and it's standardized, meaning its value always falls between -1 and +1.
- **Interpretation in this context:**
  - **Value close to +1 (Strong Positive Correlation):** A value like 0.8 or 0.9 would indicate a very strong positive linear relationship. This means that when advertising spend goes up, daily sales consistently go up in a predictable manner. The marketing team would be thrilled, as it suggests their advertising is highly effective and a good predictor of sales.
  - **Value close to -1 (Strong Negative Correlation):** A value like -0.8 or -0.9 would mean a very strong inverse linear relationship. As advertising spend increases, daily sales consistently decrease. This would be a critical finding, suggesting that the advertising might be actively harming sales or is poorly targeted.
  - **Value close to 0 (Weak or No Linear Correlation):** A value near 0 (e.g., -0.1 to 0.1) suggests a weak or non-existent linear relationship. This would tell the marketing team that simply increasing or decreasing ad spend isn't linearly impacting sales. They might need to look for other factors, different advertising channels, or non-linear relationships.
- **Advantage over Covariance:** Correlation is unitless and normalized, making it much easier to interpret and compare across different pairs of variables, regardless of their scales.

**Conclusion for the Marketing Team:**

By calculating both, especially the correlation coefficient, you can provide clear, actionable insights:

- If there's a **strong positive correlation**, you could confidently tell the marketing team that their advertising efforts are likely driving sales, and they might consider increasing their spend strategically.
- If there's a **weak or negative correlation**, you'd advise them to re-evaluate their advertising strategy, budget allocation, targeting, or even the product itself, as the current spend isn't effectively translating to sales.
- **Important Caveat:** Always remind the team that **correlation does not imply causation**. Even a strong positive correlation doesn't definitively prove that increased ad spend causes increased sales. Other factors (e.g., seasonal demand, competitor actions, economic conditions) could be at play. Further analysis, like A/B testing or regression, would be needed to establish causation.

```python
import numpy as np

advertising_spend = [200, 250, 300, 400, 500]
daily_sales = [2200, 2450, 2750, 3200, 4000]

# Convert lists to NumPy arrays
np_advertising_spend = np.array(advertising_spend)
np_daily_sales = np.array(daily_sales)

# Compute the correlation coefficient
# np.corrcoef returns a correlation matrix
correlation_matrix = np.corrcoef(np_advertising_spend, np_daily_sales)

# The correlation coefficient between the two arrays is at [0, 1] or [1, 0]
correlation_coefficient = correlation_matrix[0, 1]

print(f"Advertising Spend: {advertising_spend}")
print(f"Daily Sales: {daily_sales}")
print(f"\nCorrelation Coefficient between Advertising Spend and Daily Sales: {correlation_coefficient}")
```

#Question 9: Your team has collected customer satisfaction survey data on a scale of 1-10 and wants to understand its distribution before launching a new product.
- Explain which summary statistics and visualizations (e.g. mean, standard deviation, histogram) you’d use.
- Write Python code to create a histogram using Matplotlib for the survey data:
- survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

```python
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

survey_scores = [7, 8, 5, 9, 6, 7, 8, 9, 10, 4, 7, 6, 9, 8, 7]

# Calculate basic summary statistics for context
mean_score = np.mean(survey_scores)
median_score = np.median(survey_scores)

# Corrected Mode Calculation
counts = Counter(survey_scores)
max_count = max(counts.values())
mode_scores = [key for key, value in counts.items() if value == max_count]

std_dev = np.std(survey_scores)

print(f"Survey Scores: {survey_scores}")
print(f"Mean Score: {mean_score:.2f}")
print(f"Median Score: {median_score:.2f}")
print(f"Mode Score(s): {mode_scores}")
print(f"Standard Deviation: {std_dev:.2f}")
print("\n")

# Create the histogram
plt.figure(figsize=(10, 6))
plt.hist(survey_scores, bins=np.arange(1, 12) - 0.5, edgecolor='black', alpha=0.7)
# np.arange(1, 12) - 0.5 creates bins centered on each integer score from 1 to 10
# For example, bin for score 7 would be from 6.5 to 7.5

plt.title('Distribution of Customer Satisfaction Survey Scores (1-10)')
plt.xlabel('Satisfaction Score')
plt.ylabel('Frequency')
plt.xticks(np.arange(1, 11)) # Ensure x-axis ticks are at integer scores
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

### Explanation of the Histogram Code:

1.  **Import Libraries:**
    *   `matplotlib.pyplot as plt`: This is the standard library for creating plots in Python.
    *   `numpy as np`: Used for numerical operations, specifically for calculating standard deviation and creating appropriate bins for the histogram.

2.  **Data:**
    *   `survey_scores`: Your list of customer satisfaction scores.

3.  **Summary Statistics (for context):**
    *   The code first calculates and prints the mean, median, mode (using `plt.mlab.mode` if available, otherwise it will be empty) and standard deviation. These numerical summaries provide a quick overview before visualizing.

4.  **Create the Histogram (`plt.hist`):**
    *   `plt.figure(figsize=(10, 6))`: Sets up the plot size for better readability.
    *   `plt.hist(survey_scores, bins=np.arange(1, 12) - 0.5, edgecolor='black', alpha=0.7)`:
        *   `survey_scores`: The data to be plotted.
        *   `bins=np.arange(1, 12) - 0.5`: This is crucial for integer scores. It creates bins centered on each integer from 1 to 10. For example, a score of 7 will fall into the bin from 6.5 to 7.5, ensuring each integer score has its own distinct bar.
        *   `edgecolor='black'`: Adds black borders to the bars for better separation.
        *   `alpha=0.7`: Makes the bars slightly transparent.

5.  **Labels and Title:**
    *   `plt.title(...)`, `plt.xlabel(...)`, `plt.ylabel(...)`: These lines add a title to the plot and labels to the X and Y axes, making the plot understandable.
    *   `plt.xticks(np.arange(1, 11))`: Explicitly sets the x-axis ticks to show each integer score from 1 to 10, which is helpful since the scores are integers.

6.  **Grid and Display:**
    *   `plt.grid(axis='y', linestyle='--', alpha=0.7)`: Adds a horizontal grid for easier reading of frequencies.
    *   `plt.show()`: Displays the generated plot.
